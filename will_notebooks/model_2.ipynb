```python
import pandas as pd
```

```python
df = pd.read_csv('../targets_with_all_features.csv')
```

```python
df #preview
```

```python
df.set_index(pd.to_datetime(df['date']), inplace=True)
df.drop(columns='date', inplace=True)
df.sort_index(inplace=True)
df #review changes
```

```python
df.isnull().sum().sum() #294 nulls in total in the set
```

```python
cols = list(df.columns) #create list of all column names
```

```python
for col in cols:
    nans = df[col].isnull().sum()
    if nans > 0:
        print(col, nans) #print column name and quantity if nans are found
```

```python
#impute zeroes (no sentiment) for the pipeline and hurrican sentiment scores that are empty
df['d_f_pipeline'].fillna(0, inplace=True)
df['i_f_hurricane_storm'].fillna(0, inplace=True)
```

```python
nan_values = [] #create empty list
for col in cols: #verify changes
    nans = df[col].isnull().sum()
    if nans > 0:
        print(col, nans)
        nan_values.append(col) #add ot list
nan_values #preview list
```

```python
df.loc['05-11-2020'][nan_values]
```

```python
df.loc['05-12-2020'][nan_values]
```

these are the 2 nan value dates, a monday and tuesday in May 2020... blank for every single metric  

ideally we'd like to keep this information intact, especially with such a small quantity of rows to begin with  

I'll try imputing the average of the next 3 days for each of these (w/th/f)

```python
import numpy as np
```

```python
np.mean([df.loc['05-13-2020'][nan_values[0]], 
        df.loc['05-14-2020'][nan_values[0]], 
        df.loc['05-15-2020'][nan_values[0]]])
```

```python
for column in nan_values:
    df[column].fillna(np.mean([df.loc['05-13-2020'][column],
                         df.loc['05-14-2020'][column],
                         df.loc['05-15-2020'][column]]), inplace=True)
```

```python
df.isnull().sum().sum() #verify success
```
