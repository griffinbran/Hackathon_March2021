Lets try another OLS model - we can predict closing prices for all 10 targets  

I also need to shift some of my numbers, we don't want to have future data leaking into the predictions. Essentially all our features need to be lagged by one day so that we ensure each prediction is based on "yesterday's" numbers, since we can't possibly have today's numbers when predicting today's closing price

```python
agg_variables = list(df.columns) #all features

targets = ['phillips_66', 'bp_plc', 'valero_energy_corporation', #list of target variables
       'chevron_corporation', 'occidental_petroleum_corporation',
       'marathon_oil_corporation', 'pioneer_natural_resources_company',
       'conocophillips', 'exxon_mobil_corporation',
       'marathon_petroleum_corporation'] 

feats = ['dow_jones_transportation_average',  #list of features (not including target)
       'dow_jones_composite_average', 's&p_500',
       'dow_jones_industrial_average', 'dow_jones_utility_average',
       'workplaces', 'retail_and_recreation', 'grocery_and_pharmacy',
       'residential', 'transit_stations', 'parks', 'Value_ger_walk',
       'Value_ger_drive', 'Value_ger_trans', 'Value_ind_walk',
       'Value_ind_drive', 'Value_rus_walk', 'Value_rus_drive', 'Value_us_walk',
       'Value_us_drive', 'Value_us_trans', 'Value_jap_walk', 'Value_jap_drive',
       'Value_jap_trans', 'd_f_chevron', 'd_f_drilling', 'd_f_exxon',
       'd_f_fossil_fuel', 'd_f_marathon_oil', 'd_f_occidental_petroleum',
       'd_f_oil', 'd_f_oilfield', 'd_f_phillips_66', 'd_f_pipeline',
       'd_f_valero', 'd_g_chevron', 'd_g_drilling', 'd_g_exxon',
       'd_g_fossil_fuel', 'd_g_oil', 'd_g_oilfield', 'd_g_pipeline',
       'd_g_valero', 'i_f_airline', 'i_f_carbon_footprint', 'i_f_emissions',
       'i_f_epa', 'i_f_greenhouse', 'i_f_hurricane_storm', 'i_f_pollution',
       'i_f_sanction', 'i_f_solar', 'i_f_turbine', 'i_f_vacation',
       'i_g_airline', 'i_g_carbon_footprint', 'i_g_emissions', 'i_g_epa',
       'i_g_greenhouse', 'i_g_hurricane_storm', 'i_g_pollution',
       'i_g_sanction', 'i_g_solar', 'i_g_turbine', 'i_g_vacation']
```

```python
#loop through all features and create a column that is 1 day lagged for each
for var in agg_variables:
    df[f'lag_{var}'] = df[var].shift()
df #preview changes
```

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
df.dropna(inplace=True) #drop first row of nulls created by shift
```

```python
df.drop(columns=feats, inplace=True) #drop non-lagged columns
```

```python
df.shape #confirm shape - should be 10 target values, 10 lagged target values (1 day), 65 lagged features
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
#cycle through all 10 targets and model all of them, saving each collection of predictions
pred_list = {}
train_list = {}
test_list = {}
scores = {}
coefs = {}
for target in targets:
    X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=targets), #drop all target columns
                                                   df[target], #target is selected as y
                                                   test_size=.2, #80%/20% train/test split
                                                   shuffle=False) #predict most recent values
    X_train = sm.add_constant(X_train)
    X_test = sm.add_constant(X_test) #add intercepts
    lm = sm.OLS(y_train, X_train)
    lm_results = lm.fit()
    preds = lm_results.predict(X_test)
    train_list[target] = y_train #update dictionary with y_train
    test_list[target] = y_test #update dictionary with y_test
    pred_list[target] = preds #update dictionary with preds
    coefs[target] = lm_results.params #update dictionary with coefficient values
    scores[target] = mean_squared_error(y_test, lm_results.predict(X_test))**.5 #update dictionary with RMSE score
final_preds = pd.DataFrame(pred_list) #turn dictionary of lists into dataframe of preds
training = pd.DataFrame(train_list)
testing = pd.DataFrame(test_list)
coefs = pd.DataFrame(coefs)
final_preds.head() #preview
```

```python
def plot_preds(ytraindf, ytestdf, pred_df, title='Title', xlab=None, ylab=None):
    plt.figure(figsize=(30,20))
    for col in ytraindf.columns:
        plt.plot(ytraindf[col], color='blue') #plot each ytrain
    for col in ytestdf.columns:
        plt.plot(ytestdf[col], color='orange') #plot each ytest
    for col in pred_df.columns:
        plt.plot(pred_df[col], color='black') #plot the preds
    plt.title(title, fontsize=26)
    plt.xlabel(xlab, fontsize=20)
    plt.ylabel(ylab, fontsize=20)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)

plot_preds(training, testing, final_preds, title='Predicting Daily Closing Price - Energy Stocks',
           xlab='Year-Month', ylab='USD($)')
plt.legend(final_preds.columns, fontsize=18);
plt.savefig('../images/allfeats_preds_1.png', dpi=200, bbox_inches='tight')
```
